```python
#importing libraries
import pandas as pd
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df.head()
```

```python
X = iris_df.iloc[:, [0,1,2,3]].values
y = iris.target
```

```python
wcss = []

for i in range(1, 16):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
# Plotting the results onto a line graph
plt.plot(range(1, 16), wcss)
plt.title('elbow method')
plt.xlabel('No. of clusters')
plt.ylabel('Within cluster sum of squares') 
plt.show()
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
plt.xlabel('Sepa1 Length', fontsize=10)
plt.ylabel('Sepal Width', fontsize=10)
```

```python
plt.scatter(X[:,2], X[:,3], c=y)
plt.xlabel('Petal Length', fontsize=10)
plt.ylabel('Petal Width', fontsize=10)
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(X)
```

```python
#center points of the data
center = kmeans.cluster_centers_
print(center)
```

```python
#gives the cluster to which the data observations belong.
new_labels = kmeans.labels_
```

```python
# Plotting the cluster
fig, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].scatter(X[:, 0], X[:, 1], c=y, cmap='gist_rainbow',
edgecolor='k', s=150)
axes[1].scatter(X[:, 0], X[:, 1], c=new_labels, cmap='jet',
edgecolor='k', s=150)
axes[0].set_xlabel('Sepal length', fontsize=12)
axes[0].set_ylabel('Sepal width', fontsize=12)
axes[1].set_xlabel('Sepal length', fontsize=12)
axes[1].set_ylabel('Sepal width', fontsize=12)
axes[0].tick_params(direction='in', length=10, width=5, labelsize=10)
axes[1].tick_params(direction='in', length=10, width=5, labelsize=10)
axes[0].set_title('Actual', fontsize=12)
axes[1].set_title('Predicted', fontsize=12)
```
